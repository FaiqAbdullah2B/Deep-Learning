# Task 02
All tasks have been implemented and are ready to run. We use a 3-layer neural network (Input, Hidden, Output) to classify the `make_moons` dataset.

```python
import numpy as np
import pandas as pd
import sklearn
import sklearn.datasets
import matplotlib.pyplot as plt

np.random.seed(0)


X, y = sklearn.datasets.make_moons(200, noise=0.20)

plt.figure(figsize=(8,6))
plt.scatter(X[:,0], X[:,1], s=40, c=y, cmap=plt.cm.Spectral)
plt.title('Generated make_moons Dataset')
plt.show()
```

### Global Hyperparameters & Helper Functions
Setting up the dimensions, learning rate (`epsilon`), and regularization (`reg_lambda`). We also define the `calculate_loss`, `predict`, and decision boundary plotting functions here.

```python
num_examples = len(X) # training set size
nn_input_dim = 2 # input layer dimensionality
nn_output_dim = 2 # output layer dimensionality

# Gradient descent parameters
epsilon = 0.01 # learning rate for gradient descent
reg_lambda = 0.01 # regularization strength

# ii. function to plot a decision boundary
def plot_decision_boundary(pred_func):
    x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
    y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
    h = 0.01
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    Z = pred_func(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, cmap=plt.cm.Spectral, alpha=0.8)
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Spectral)

# iv. Function to evaluate the total loss on dataset
def calculate_loss(model):
    W1, b1, W2, b2 = model['W1'], model['b1'], model['W2'], model['b2']
    # Forward propagation
    z1 = X.dot(W1) + b1
    a1 = np.tanh(z1)
    z2 = a1.dot(W2) + b2
    exp_scores = np.exp(z2)
    probs = exp_scores / np.sum(exp_scores, axis=1, keepdims=True)
    # Calculating the loss
    corect_logprobs = -np.log(probs[range(num_examples), y])
    data_loss = np.sum(corect_logprobs)
    # Add regulatization term to loss
    data_loss += reg_lambda/2 * (np.sum(np.square(W1)) + np.sum(np.square(W2)))
    return 1./num_examples * data_loss

# vi. Prediction function
def predict(model, x):
    W1, b1, W2, b2 = model['W1'], model['b1'], model['W2'], model['b2']
    z1 = x.dot(W1) + b1
    a1 = np.tanh(z1)
    z2 = a1.dot(W2) + b2
    exp_scores = np.exp(z2)
    probs = exp_scores / np.sum(exp_scores, axis=1, keepdims=True)
    return np.argmax(probs, axis=1)
```

## Task 01: Standard Model (Batch Gradient Descent)
Implementing forward propagation, backpropagation, and weight updates.

```python
def build_model(nn_hdim, num_passes=20000, print_loss=False):
    np.random.seed(0)
    W1 = np.random.randn(nn_input_dim, nn_hdim) / np.sqrt(nn_input_dim)
    b1 = np.zeros((1, nn_hdim))
    W2 = np.random.randn(nn_hdim, nn_output_dim) / np.sqrt(nn_hdim)
    b2 = np.zeros((1, nn_output_dim))
    
    model = {}
    
    for i in range(0, num_passes):
        # Forward propagation
        z1 = X.dot(W1) + b1
        a1 = np.tanh(z1)
        z2 = a1.dot(W2) + b2
        exp_scores = np.exp(z2)
        probs = exp_scores / np.sum(exp_scores, axis=1, keepdims=True)
        
        # Back propagation
        delta3 = probs
        delta3[range(num_examples), y] -= 1
        
        dW2 = (a1.T).dot(delta3)
        db2 = np.sum(delta3, axis=0, keepdims=True)
        
        delta2 = delta3.dot(W2.T) * (1 - np.power(a1, 2))
        dW1 = np.dot(X.T, delta2)
        db1 = np.sum(delta2, axis=0)
        
        # Add regularization terms
        dW2 += reg_lambda * W2
        dW1 += reg_lambda * W1
        
        # Parameter update
        W1 += -epsilon * dW1
        b1 += -epsilon * db1
        W2 += -epsilon * dW2
        b2 += -epsilon * db2
        
        model = {'W1': W1, 'b1': b1, 'W2': W2, 'b2': b2}
        
        if print_loss and i % 1000 == 0:
            print(f'Loss after iteration {i}: {calculate_loss(model):.5f}')
            
    return model

# Train and plot standard model
print('Training Task 1 Model...')
model = build_model(3, num_passes=10000, print_loss=True)
plt.figure(figsize=(8,6))
plot_decision_boundary(lambda x: predict(model, x))
plt.title('Decision Boundary - Standard Batch Gradient Descent')
plt.show()
```

## Task 02 i & ii: Minibatch Gradient Descent & Learning Rate Decay
Instead of updating weights based on the whole dataset at once, we chunk it into mini-batches. We also implement two forms of learning rate decay: `time` (inverse scaling) and `exponential`.

```python
def build_model_minibatch(nn_hdim, epochs=5000, batch_size=32, initial_lr=0.01, decay_type='time', print_loss=False):
    np.random.seed(0)
    W1 = np.random.randn(nn_input_dim, nn_hdim) / np.sqrt(nn_input_dim)
    b1 = np.zeros((1, nn_hdim))
    W2 = np.random.randn(nn_hdim, nn_output_dim) / np.sqrt(nn_hdim)
    b2 = np.zeros((1, nn_output_dim))
    
    model = {}
    decay_rate = 0.001
    
    for epoch in range(epochs):
        # Apply Learning Rate Decay
        if decay_type == 'time':
            lr = initial_lr / (1. + decay_rate * epoch)
        elif decay_type == 'exponential':
            lr = initial_lr * np.exp(-decay_rate * epoch)
        else:
            lr = initial_lr
            
        # Shuffle data for mini-batches
        indices = np.arange(num_examples)
        np.random.shuffle(indices)
        X_shuffled = X[indices]
        y_shuffled = y[indices]
        
        for start_idx in range(0, num_examples, batch_size):
            end_idx = min(start_idx + batch_size, num_examples)
            X_batch = X_shuffled[start_idx:end_idx]
            y_batch = y_shuffled[start_idx:end_idx]
            batch_len = end_idx - start_idx
            
            # Forward propagation
            z1 = X_batch.dot(W1) + b1
            a1 = np.tanh(z1)
            z2 = a1.dot(W2) + b2
            exp_scores = np.exp(z2)
            probs = exp_scores / np.sum(exp_scores, axis=1, keepdims=True)
            
            # Back propagation
            delta3 = probs
            delta3[range(batch_len), y_batch] -= 1
            
            dW2 = (a1.T).dot(delta3)
            db2 = np.sum(delta3, axis=0, keepdims=True)
            
            delta2 = delta3.dot(W2.T) * (1 - np.power(a1, 2))
            dW1 = np.dot(X_batch.T, delta2)
            db1 = np.sum(delta2, axis=0)
            
            dW2 += reg_lambda * W2
            dW1 += reg_lambda * W1
            
            W1 += -lr * dW1
            b1 += -lr * db1
            W2 += -lr * dW2
            b2 += -lr * db2
            
        model = {'W1': W1, 'b1': b1, 'W2': W2, 'b2': b2}
        
        if print_loss and epoch % 1000 == 0:
            print(f'Epoch {epoch} | LR: {lr:.5f} | Loss: {calculate_loss(model):.5f}')
            
    return model

print('Training Minibatch Model with Exponential Decay...')
mb_model = build_model_minibatch(3, epochs=5000, batch_size=32, decay_type='exponential', print_loss=True)
plt.figure(figsize=(8,6))
plot_decision_boundary(lambda x: predict(mb_model, x))
plt.title('Decision Boundary - Minibatch & Exp Decay')
plt.show()
```

## Task 02 iii: Model Capacity (Different Hidden Neurons)
We train the model with 1, 3, and 10 hidden neurons. 
* **1 Neuron:** Underfits. It acts essentially linearly.
* **3 Neurons:** Good fit. It has enough capacity to shape around the clusters.
* **10 Neurons:** Starts to overfit, creating overly complex boundaries around noise.

```python
plt.figure(figsize=(16, 12))
hidden_layer_dimensions = [1, 3, 10]

for i, nn_hdim in enumerate(hidden_layer_dimensions):
    plt.subplot(2, 2, i+1)
    plt.title(f'Hidden Layer size {nn_hdim}')
    print(f'Training model with {nn_hdim} hidden neurons...')
    model = build_model(nn_hdim, num_passes=10000, print_loss=False)
    plot_decision_boundary(lambda x: predict(model, x))

plt.tight_layout()
plt.show()
```

## Task 02 iv: Two Hidden Layers
Extending the model architecture to include a second hidden layer. This requires expanding both forward and backpropagation to calculate gradients across `W1`, `W2`, and `W3`.

```python
def predict_2hl(model, x):
    W1, b1, W2, b2, W3, b3 = model['W1'], model['b1'], model['W2'], model['b2'], model['W3'], model['b3']
    z1 = x.dot(W1) + b1
    a1 = np.tanh(z1)
    z2 = a1.dot(W2) + b2
    a2 = np.tanh(z2)
    z3 = a2.dot(W3) + b3
    exp_scores = np.exp(z3)
    probs = exp_scores / np.sum(exp_scores, axis=1, keepdims=True)
    return np.argmax(probs, axis=1)

def build_model_2_hidden_layers(hdim1=5, hdim2=5, num_passes=10000):
    np.random.seed(0)
    W1 = np.random.randn(nn_input_dim, hdim1) / np.sqrt(nn_input_dim)
    b1 = np.zeros((1, hdim1))
    W2 = np.random.randn(hdim1, hdim2) / np.sqrt(hdim1)
    b2 = np.zeros((1, hdim2))
    W3 = np.random.randn(hdim2, nn_output_dim) / np.sqrt(hdim2)
    b3 = np.zeros((1, nn_output_dim))
    
    model = {}
    
    for i in range(0, num_passes):
        # Forward prop
        z1 = X.dot(W1) + b1
        a1 = np.tanh(z1)
        z2 = a1.dot(W2) + b2
        a2 = np.tanh(z2)
        z3 = a2.dot(W3) + b3
        exp_scores = np.exp(z3)
        probs = exp_scores / np.sum(exp_scores, axis=1, keepdims=True)
        
        # Back prop
        delta4 = probs
        delta4[range(num_examples), y] -= 1
        
        dW3 = (a2.T).dot(delta4)
        db3 = np.sum(delta4, axis=0, keepdims=True)
        
        delta3 = delta4.dot(W3.T) * (1 - np.power(a2, 2))
        dW2 = np.dot(a1.T, delta3)
        db2 = np.sum(delta3, axis=0)
        
        delta2 = delta3.dot(W2.T) * (1 - np.power(a1, 2))
        dW1 = np.dot(X.T, delta2)
        db1 = np.sum(delta2, axis=0)
        
        dW3 += reg_lambda * W3
        dW2 += reg_lambda * W2
        dW1 += reg_lambda * W1
        
        W1 += -epsilon * dW1
        b1 += -epsilon * db1
        W2 += -epsilon * dW2
        b2 += -epsilon * db2
        W3 += -epsilon * dW3
        b3 += -epsilon * db3
        
    return {'W1': W1, 'b1': b1, 'W2': W2, 'b2': b2, 'W3': W3, 'b3': b3}

print('Training 2-Hidden-Layer Model...')
model_2hl = build_model_2_hidden_layers(hdim1=5, hdim2=5)
plt.figure(figsize=(8,6))
plot_decision_boundary(lambda x: predict_2hl(model_2hl, x))
plt.title('Decision Boundary - 2 Hidden Layers (5x5)')
plt.show()
```
